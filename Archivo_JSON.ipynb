```python
import json
```

```python
datos = {
  "nombre": "Eréndira",
  "materia": "base de datos NSQL",
  "alumnos": [
    {"nombre": "Andrea Itzel", "edad": 19, "Matricula":"24102091-1", "email":"Itzelfalcon520@gmail.com"},
    {"nombre": "Geovani", "edad": 20, "Matricula":"20102165-1", "email":"geovani.ortc@gmail.com"},
    {"nombre": "Luz María", "edad": 20, "Matricula":"24109021-1", "email":"Geovani.ortc@gmail.com"},
    {"nombre": "Marco", "edad": 33, "Matricula":"24106761-1", "email":"marco.aram922@gmail.com"}
  ]
}

# "alumnos" es una lista
# "Materia" es una cadena (Str)
# "nombre" es una cadena (Str)
# "edad" es un número (Int)
# "Matricula" es una cadena, pues mezcla números y caracteres especiales (Str)
# "email" es una cadena (Str)
```

```python
json_texto = json.dumps(datos, indent=4, ensure_ascii=False)
print(json_texto)
```

```python
with open("alumnos.json", "w", encoding="utf-8") as f:
    json.dump(datos, f, indent=4, ensure_ascii=False)
from google.colab import files
files.download("alumnos.json")
```
